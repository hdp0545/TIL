```python
# 인터파크 : 뭉크 리뷰 데이터 스크래핑하기
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
import numpy as np
import time
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://tickets.interpark.com/goods/24005098")
```

```python
def make_df():
    titles = driver.find_elements(By.CSS_SELECTOR, "strong.bbsTitleText")
    strTitles = [] 
    for title in titles:
        strTitles.append(title.text.strip())
    
    contents = driver.find_elements(By.CSS_SELECTOR, "p.bbsText")
    strContents = []
    for content in contents:
        strContents.append(content.text.strip())
    
    df = pd.DataFrame({"제목" : strTitles,
                       "내용" : strContents})
    return df
```

```python


total = []

while True:
    liList = driver.find_elements(By.CSS_SELECTOR, "ol.pageNumWrap>li")
    for li in liList:
        li.click()
        time.sleep(0.5)
        df = make_df()
        total.append(df)
    try:
        nextBtn = driver.find_element(By.CSS_SELECTOR, "a.pageNextBtn.pageArrow")
        nextBtn.click()
        time.sleep(0.5)
    except:
        break
```

```python
totalDf = pd.concat(total, ignore_index=True)
review = totalDf
```

```python
review["TEXT"] = review["제목"] + " " + review["내용"]
review
```

```python
words = []
for i in review["TEXT"]:
    words += i.split()
len(words)
```

```python
dic = {}

for word in words:
    if word in dic:
        dic[word] += 1
    else:
        dic[word] = 1

result = pd.DataFrame({"단어": dic.keys()
                     , "빈도수" : dic.values()})
result.sort_values("빈도수", ascending=False)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
# 문제집 : 길이, 무게
# 정답지 : 생선의 종류

fish = pd.read_csv("fish.csv")
fish
```

```python
fish2 = pd.read_csv("fish2.csv")
fish2
```

```python
X = fish2[['Weight', 'Length','Diagonal', 'Height', 'Width']].to_numpy()
Y = fish2['Species'].to_numpy()

train_x, test_x, train_y, test_y = train_test_split(X, Y)
```

```python
# axis=0이면 행방향
mean = train_x.mean(axis=0)
std = train_x.std(axis=0)
mean, std

train_scaled = (train_x - mean) / std
test_scaled = (test_x - mean) / std #표본평균과 표본표준편차를 빼줘야 함.
```

```python
model = KNeighborsClassifier()
model.fit(train_scaled, train_y)

box = [0] * 30
for _ in range(100):
    train_x, test_x, train_y, test_y = train_test_split(X, Y)
    # axis=0이면 행방향
    mean = train_x.mean(axis=0)
    std = train_x.std(axis=0)
    mean, std
    
    train_scaled = (train_x - mean) / std
    test_scaled = (test_x - mean) / std #표본평균과 표본표준편차를 빼줘야 함.
    for n in range(1, 31):
        model.n_neighbors = n
        score = model.score(test_scaled, test_y)
        box[n-1] += score

# model.score(test_scaled, test_y)
```

```python
plt.plot(range(1,31), box)
plt.show()
```
